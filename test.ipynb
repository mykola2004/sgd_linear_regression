```python
import numpy as np
from sgd import SGDLinearRegressor
```

Let's create artificial independent features 

```python
x1 = np.random.uniform(low=-1, high=1, size=1000)
x2 = np.random.uniform(low=-5, high=5, size=1000)
```

As well we will need white noise for creation of dependent feature

```python
epsilon = np.random.normal(loc=0, scale=1, size=1000)
```

Finally, let's define the dependent feature as a linear combination of independent features plus previously defined white noise

```python
y = 10*x1 + 0.1*x2 + epsilon
```

We will need stacked together vectors x1 and x2

```python
X = np.stack((x1, x2), axis=1)
```

Here will be fitted to the data linear regression model from the sgd.py file

```python
reg_sgd = SGDLinearRegressor(0.1, regularization=0, delta_converged=0.00000001, max_steps=4000, batch_size=300)
reg_sgd.fit(X, y)
print("Weights: ", reg_sgd.get_weights_and_bias())
```
